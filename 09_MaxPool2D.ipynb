## Maxpooling 이란?

 maxPool2D.png

> 84개의 출력 * (784개의 가중치 + 1개의 bias) = 65940개의 가중치가 히든레이어를 구성하는데 필요하다

> 10개의 수식 * (84개의 가중치 + 1개의 bias) = 850개의 가중치가 마지막 출력층을 구성하는데 필요하다

maxPool2D_1.png

> 84개의 출력 * (2400개의 가중치 + 1개의 bias) = 201,684개의 가중치가 히든레이어 구성에 필요하다.

> Convolution layer로 출력된 특징 맵 갯수(20,20,6)의 증가 -> Flatten 이후 입력으로 사용될 column 수의 증가 -> 컴퓨터가 찾아야하는 가중치 수의 증가

> 이를 위해 도입된 방법이 **Pooling**

> Pooling은 Flatten layer 이후에 사용된 가중치의 수를 작게 유지하기 위해 입력으로 사용할 column수를 조절하는 것이 목적임

maxPool2D_2.png

> Convolution layer로 출력된 특징 맵의 사이즈가 maxpool layer를 거치면서 절반으로 줄어듬을 확인가능 

> (24, 24, 3) -> (12, 12, 3) ,  (8, 8, 6) -> (4, 4, 6)

## Maxpooling 원리

> MaxPooling은 (2n, 2n) 사이즈의 이미지를 (n, n) 사이즈로 이미지를 줄이는 일이다.

> 가로 세로 (2,2) 영역안에서 가장 큰 수를 남기면 된다
