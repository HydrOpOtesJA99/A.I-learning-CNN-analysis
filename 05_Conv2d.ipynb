## 2. Convolution layer 추가한 코드의 확인

Conv2d_2.png





*   [28, 28, 1]: Convolution 연산은 3차원 형태의 관측치를 input으로 받는다(개발자가 정함, 개발자 맘대로)

*   ~.reshape(60000, 28, 28, 1): 위의 이유 때문에 독립변수를 reshape가 필요
  * 컬러 이미지는 3차원 형태이므로 흑백 이미지도 3차원 형태로 바꿔준다고 생각하면 될듯

*   3, 6: 은 필터셋의 갯수

*   kernal_size=5: 필터셋의 크기, (5,5)사이즈의 필터셋을 사용하기 위해 5라고 적음

*   3채널, 6채널의 특징맵: 
  * 첫 번째 convolution layer는 3개의 필터셋을 사용하므로 3개의 특징맵을 만든다
  * 두 번째 convolution layer는 6개의 필터셋을 사용하므로 6개의 특징맵을 만든다 

*   Flatten()(H): 6채널의 특징맵 모두를 flatten layer로 픽셀단위로 한줄로 펼쳐서 학습한다
