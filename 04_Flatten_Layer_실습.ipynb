# 4_Flatten_Layer를_활용한_이미지_학습



```
import tensorflow as tf
import pandas as pd
```

## with reshape

### 1)과거의 데이터 준비

```
#1. 과거의 데이터를 준비한다.

(독립, 종속), _ = tf.keras.datasets.mnist.load_data()
print(독립.shape, 종속.shape) # -> print 결과: 값이 표가 아니므로 표로 바꿔줘야 함

독립 = 독립.reshape(60000,784)
종속 = pd.get_dummies(종속)
print(독립.shape, 종속.shape) # -> print 결과: 행, 열을 가진 '표'형 데이터 생성
```

```
종속.head()
```

```
종속.tail()
```



*   (60000, 28, 28) --**reshape**-> (60000, 784)
*   28 x 28(행 x 열)을 784개의 column으로 바꿔줌
---
*   (60000,) --**원핫인코딩; pd.get_dummies**-> (60000, 10)
*   0~9까지 10개의 범주를 10개의 column으로 바꿔줌


### 2) 모델의 구조 만들기

```
#2. 모델의 구조를 만든다.

X = tf.keras.layers.Input(shape=[784]) # 독립변수 784개
H = tf.keras.layers.Dense(84, activation='swish')(X) # 84개의 은닉층 추가
Y = tf.keras.layers.Dense(10, activation='softmax')(H) # 종속변수 10개
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

### 3) 데이터로 모델 학습

```
#3. 데이터로 모델을 학습(FIT)한다.

model.fit(독립, 종속, epochs=10)
```

### 4) 모델을 이용

```
#4. 모델을 이용한다.

pred = model.predict(독립[0:5])
pd.DataFrame(pred).round(2) # DataFrame의 형태로 변형해 출력, round(2) -> 소수점 두번째 자리에서 반올림
```

```
# 정답지

종속[0:5] # 위와 비교해보니 정확하게 예측된듯 하다.
```

## with flatten

### 1)과거의 데이터 준비

```
#1. 과거의 데이터를 준비한다.

(독립, 종속), _ = tf.keras.datasets.mnist.load_data()
print(독립.shape, 종속.shape) # -> print 결과: 값이 표가 아니므로 표로 바꿔줘야 함

# 독립 = 독립.reshape(60000,784) # -> flatten 사용하므로 reshape 사용 X
종속 = pd.get_dummies(종속)
print(독립.shape, 종속.shape) # -> print 결과: 행, 열을 가진 '표'형 데이터 생성
```

```
종속.head()
```

```
종속.tail()
```



*   (60000, 28, 28) --**reshape**-> (60000, 784)
*   28 x 28(행 x 열)을 784개의 column으로 바꿔줌
---
*   (60000,) --**원핫인코딩; pd.get_dummies**-> (60000, 10)
*   0~9까지 10개의 범주를 10개의 column으로 바꿔줌


### 2) 모델의 구조 만들기

```
#2. 모델의 구조를 만든다.

X = tf.keras.layers.Input(shape=[28, 28]) # reshape 사용X, 따라서 [28, 28]로 적음

H = tf.keras.layers.Flatten()(X) # Flatten을 이용해서 X data를 한줄로 펴준다
H = tf.keras.layers.Dense(84, activation='swish')(H) 

Y = tf.keras.layers.Dense(10, activation='softmax')(H) # 종속변수 10개

model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

### 3) 데이터로 모델 학습

```
#3. 데이터로 모델을 학습(FIT)한다.

model.fit(독립, 종속, epochs=10)
```

### 4) 모델을 이용

```
#4. 모델을 이용한다.

pred = model.predict(독립[-5:]) # [-5:] -> 뒤에서부터 5개의 값을 print하라
pd.DataFrame(pred).round(2) # DataFrame의 형태로 변형해 출력, round(2) -> 소수점 두번째 자리에서 반올림
```

```
# 정답지
# [-5:] -> 뒤에서부터 5개의 값을 print하라

종속[-5:] # 위와 비교해보니 정확하게 예측된듯 하다. 
```
