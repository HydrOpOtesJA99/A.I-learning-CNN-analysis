## 1. 필터의 이해



>  **필터셋**: (5,5)의 필터가 채널수 만큼 겹쳐진 것이므로 '필터셋'이라고 표현, 실제로는 그냥 '필터'라고 부름

filter_1.png



*   3차원 형태로 된 가중치: F1, F2, F3 중 개별 필터셋 하나가 3차원 형태것(5, 5, ?)라는 것 이고 이 필터셋이 여러 채널 존재하는 것이다.

*   "특징맵" 전체를 본다: **흑백**의 경우 채널이 하나이므로 필터셋의 형태는 (5, 5, 1)이 된다.



filter.png

*   "특징맵" 전체를 본다: **컬러**의 경우 채널이 3개이므로 필터셋의 형태는 (5, 5, 3)이 된다.

filter_2.png



*   (3, 5, 5, ?), (6, 5, 5, ?): 따라서 필터셋 전체의 형태는 4차원 형태가 된다



filter_3.png



1.   흑백이라 값이 1인(28, 28, 1) 입력 이미지

2.   (5, 5, 1) 필터 3개를 가진 필터셋 준비됨

3.   위 필터셋으로 3채널의 특징맵이 생성된다(3개의 특징 맵이 생성된다)
  >   필터의 사이즈가 5 이므로 크기가 28 --> 24 로 줄어든 특징맵 생성

  >   (입력이미지 크기(28) - 필터 크기(5) + 1) = 24

4.   두 번째 convolution layer에서는 6장의 필터를 준비하였고 필터셋의 채널 수는 앞선 레이어의 결과와 같은 값인 **3**으로 되어있다. 

5.   (5, 5, 3) 필터 6개를 가진 필터셋 준비되었다
  >  새로운 특징맵(20, 20, 6)은 앞의 특징맵(24, 24, 3)을 보고 생성되었다.

  >  (입력이미지 크기 (24) - 필터 크기(5) + 1) = 20

> ** Convolution layer의 Convolution filter를 뜯어보면 가중치들로 이루어져 있고 컴퓨터는 특징 맵을 만들어 내는 이 convolution filter를 학습하는 것이다. **
