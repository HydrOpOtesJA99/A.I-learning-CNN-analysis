LeNet.png

1. INPUT 32x32: 입력으로 32 x 32 사용

2. Feature Map (6 @ 28 x 28): filter_size = 5를 사용하여 28 x 28(=32-5+1), 6개의 특징맵 추출

3. MaxPooling (6 @ 14 x 14): 28 -> 14 MaxPooling

4. Feature Map (16 @ 10 x 10): filter_size = 5를 사용하여 10 x 10(=14-5+1), 6개의 특징맵 추출 

5. MaxPooling (16 @ 5 x 5): 10 -> 5 MaxPooling

6. Flatten: 16 x 5 x 5 = 400개의 column으로 펼침

7. 120개의 node, 84개의 node를 가진 hidden layer 추가

8. 마지막으로 10개의 출력을 만듦
