## 2. 포함 관계의 관점에서 차원

배열의 깊이.png

```
x1 = np.array([5.1, 3.5, 1.4, 0.2])
print(x1.ndim, x1.shape)

# 1차원 형태, (4,)
```
*   위 'x1'의 shape는 **4개의 요소(5.1, 3.5, 1.4, 0.2)**를 지니는 
*   **1차원(숫자가 '4', 1개 적혀있음) 깊이를 가진다고 할 수 있다.**

```
x1 = np.array([5.1, 3.5, 1.4, 0.2])
x2 = np.array([4.9, 3.0, 1.4, 0.2])
x3 = np.array([4.7, 3.2, 1.3, 0.2])

아이리스 = np.array([x1, x2, x3])
print(아이리스.ndim, 아이리스.shape)

# 2차원 형태, (3,4)
```

*   위 '아이리스'의 shape는 **3개의 요소(x1, x2, x3)**를 지니고 
*   그 속에 **4개의 요소(5.1, 3.5, 1.4, 0.2)**를 가진다 
*   **즉 2차원의 깊이를 가진다고 할 수 있다.**

배열의 깊이_1.png



```
d4 = np.array([d3, d3])
print(d4.ndim, d4.shape)

# 4차원 형태, (2, 4, 5, 3)
```
*   위 'd4'의 shape는 
*   2개의 요소(d3, d3) 속, 
*   4개의 요소(d2, d2, d2, d2) 속, 
*   5개의 요소(d1, d1, d1, d1, d1) 속, 
*   3개의 요소(1, 2, 3)을 가진다
*   **즉 4차원의 깊이를 가진다고 할 수 있다.**
